## Calculate Customer Visit Frequency

### Subtask:
Group the data by 'Member_number', sort purchases by date for each customer, and calculate the average time difference between consecutive visits to determine customer visit frequency.


**Reasoning**:

```python
customer_visit_frequency_days = time_diffs.groupby(level='Member_number').mean()

print("Average visit frequency per customer (in days, first 10 values):")
print(customer_visit_frequency_days.head(10))
```

## Customer Segmentation based on Visit Patterns

### Subtask:
Segment customers based on their visit frequency and favorite shopping days to identify distinct customer groups.


**Reasoning**:
First, I will convert the `customer_visit_frequency_days` Series to a DataFrame, reset its index, and rename the column to 'avg_days_between_visits'. Then, I will merge it with the `favorite_shopping_day_per_customer` DataFrame on 'Member_number' to create the `customer_segmentation_df`, which aligns with the first instruction.



```python
customer_visit_frequency_df = customer_visit_frequency_days.reset_index()
customer_visit_frequency_df.rename(columns={'Date': 'avg_days_between_visits'}, inplace=True)

customer_segmentation_df = pd.merge(customer_visit_frequency_df,
                                    favorite_shopping_day_per_customer[['Member_number', 'DayOfWeek']],
                                    on='Member_number',
                                    how='left')

print("First 5 records of customer_segmentation_df:")
display(customer_segmentation_df.head())
```

**Reasoning**:
To handle missing values, I will first check if there are any NaN values in the 'avg_days_between_visits' column and fill them with the mean of that column. Then, I will check for any remaining missing values across the DataFrame and drop those rows to ensure data completeness for clustering.



```python
print(f"Missing values before handling:\n{customer_segmentation_df.isnull().sum()}\n")

# Fill missing values in 'avg_days_between_visits' with the mean
if customer_segmentation_df['avg_days_between_visits'].isnull().any():
    mean_avg_days = customer_segmentation_df['avg_days_between_visits'].mean()
    customer_segmentation_df['avg_days_between_visits'].fillna(mean_avg_days, inplace=True)
    print(f"Filled missing 'avg_days_between_visits' with mean: {mean_avg_days:.2f}\n")

# Drop rows with any remaining missing values (e.g., if 'DayOfWeek' had NaNs)
initial_rows = customer_segmentation_df.shape[0]
customer_segmentation_df.dropna(inplace=True)
rows_dropped = initial_rows - customer_segmentation_df.shape[0]
if rows_dropped > 0:
    print(f"Dropped {rows_dropped} rows with remaining missing values.\n")

print(f"Missing values after handling:\n{customer_segmentation_df.isnull().sum()}\n")
print("First 5 records of customer_segmentation_df after handling missing values:")
display(customer_segmentation_df.head())
```

**Reasoning**:
The previous code generated a `FutureWarning` due to the use of `inplace=True` with `fillna` on a potentially copied slice. To avoid this warning and ensure consistent behavior, I will explicitly reassign the result of `fillna` back to the 'avg_days_between_visits' column. Additionally, I will proceed with the next steps of the subtask: converting 'DayOfWeek' to numerical format using one-hot encoding, standardizing numerical features, determining the optimal number of clusters using the silhouette score for K-Means, and then applying K-Means to segment customers.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

print(f"Missing values before handling:\n{customer_segmentation_df.isnull().sum()}\n")

# Fill missing values in 'avg_days_between_visits' with the mean
if customer_segmentation_df['avg_days_between_visits'].isnull().any():
    mean_avg_days = customer_segmentation_df['avg_days_between_visits'].mean()
    customer_segmentation_df['avg_days_between_visits'] = customer_segmentation_df['avg_days_between_visits'].fillna(mean_avg_days)
    print(f"Filled missing 'avg_days_between_visits' with mean: {mean_avg_days:.2f}\n")

# Drop rows with any remaining missing values (e.g., if 'DayOfWeek' had NaNs)
initial_rows = customer_segmentation_df.shape[0]
customer_segmentation_df.dropna(inplace=True)
rows_dropped = initial_rows - customer_segmentation_df.shape[0]
if rows_dropped > 0:
    print(f"Dropped {rows_dropped} rows with remaining missing values.\n")

print(f"Missing values after handling:\n{customer_segmentation_df.isnull().sum()}\n")

# 3. Convert 'DayOfWeek' to numerical format using one-hot encoding
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
day_of_week_encoded = encoder.fit_transform(customer_segmentation_df[['DayOfWeek']])
day_of_week_df = pd.DataFrame(day_of_week_encoded, columns=encoder.get_feature_names_out(['DayOfWeek']), index=customer_segmentation_df.index)

# Combine numerical and encoded features
customer_features = pd.concat([
    customer_segmentation_df[['avg_days_between_visits']],
    day_of_week_df
], axis=1)

print("First 5 records of customer_features after one-hot encoding:")
display(customer_features.head())

# 4. Standardize the numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_features)

print("Shape of scaled_features:", scaled_features.shape)

# 5. Determine the optimal number of clusters for K-Means using silhouette score
# We will test a range of clusters, e.g., from 2 to 10
max_clusters = min(10, len(customer_segmentation_df) - 1) # Ensure max_clusters doesn't exceed number of samples
if max_clusters < 2: # Handle cases where there are too few samples for clustering
    print("Not enough samples to perform clustering.")
    # Fallback or exit strategy if clustering is not possible
    optimal_clusters = 1 # Or handle this as an error
else:
    silhouette_scores = []
    K_range = range(2, max_clusters + 1)

    for n_clusters in K_range:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(scaled_features)
        if len(set(cluster_labels)) > 1: # Silhouette score requires at least 2 clusters
            silhouette_scores.append(silhouette_score(scaled_features, cluster_labels))
        else:
            silhouette_scores.append(-1) # Assign a low score if only one cluster is formed

    if silhouette_scores:
        optimal_clusters = K_range[silhouette_scores.index(max(silhouette_scores))]
        print(f"Optimal number of clusters (based on Silhouette Score): {optimal_clusters}")
    else:
        optimal_clusters = 2 # Default to 2 if no valid silhouette scores are found
        print("Could not determine optimal clusters using Silhouette Score, defaulting to 2.")

# Apply K-Means with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
customer_segmentation_df['Cluster'] = kmeans.fit_predict(scaled_features)

# 6. Display the head of customer_segmentation_df with the new 'Cluster' column
print("\nFirst 5 records of customer_segmentation_df with 'Cluster' column:")
display(customer_segmentation_df.head())

# Display the distribution of customers across the identified clusters
print("\nDistribution of customers across clusters:")
print(customer_segmentation_df['Cluster'].value_counts())

plt.figure(figsize=(8, 5))
sns.countplot(data=customer_segmentation_df, x='Cluster', palette='viridis')
plt.title('Distribution of Customers Across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` because `palette` was used without `hue` in `sns.countplot`. To fix this and follow the recommended practice, I will explicitly assign the `Cluster` column to `hue` and set `legend=False` in the `sns.countplot` call.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

print(f"Missing values before handling:\n{customer_segmentation_df.isnull().sum()}\n")

# Fill missing values in 'avg_days_between_visits' with the mean
if customer_segmentation_df['avg_days_between_visits'].isnull().any():
    mean_avg_days = customer_segmentation_df['avg_days_between_visits'].mean()
    customer_segmentation_df['avg_days_between_visits'] = customer_segmentation_df['avg_days_between_visits'].fillna(mean_avg_days)
    print(f"Filled missing 'avg_days_between_visits' with mean: {mean_avg_days:.2f}\n")

# Drop rows with any remaining missing values (e.g., if 'DayOfWeek' had NaNs)
initial_rows = customer_segmentation_df.shape[0]
customer_segmentation_df.dropna(inplace=True)
rows_dropped = initial_rows - customer_segmentation_df.shape[0]
if rows_dropped > 0:
    print(f"Dropped {rows_dropped} rows with remaining missing values.\n")

print(f"Missing values after handling:\n{customer_segmentation_df.isnull().sum()}\n")

# 3. Convert 'DayOfWeek' to numerical format using one-hot encoding
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
day_of_week_encoded = encoder.fit_transform(customer_segmentation_df[['DayOfWeek']])
day_of_week_df = pd.DataFrame(day_of_week_encoded, columns=encoder.get_feature_names_out(['DayOfWeek']), index=customer_segmentation_df.index)

# Combine numerical and encoded features
customer_features = pd.concat([
    customer_segmentation_df[['avg_days_between_visits']],
    day_of_week_df
], axis=1)

print("First 5 records of customer_features after one-hot encoding:")
display(customer_features.head())

# 4. Standardize the numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_features)

print("Shape of scaled_features:", scaled_features.shape)

# 5. Determine the optimal number of clusters for K-Means using silhouette score
# We will test a range of clusters, e.g., from 2 to 10
max_clusters = min(10, len(customer_segmentation_df) - 1) # Ensure max_clusters doesn't exceed number of samples
if max_clusters < 2: # Handle cases where there are too few samples for clustering
    print("Not enough samples to perform clustering.")
    # Fallback or exit strategy if clustering is not possible
    optimal_clusters = 1 # Or handle this as an error
else:
    silhouette_scores = []
    K_range = range(2, max_clusters + 1)

    for n_clusters in K_range:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(scaled_features)
        if len(set(cluster_labels)) > 1: # Silhouette score requires at least 2 clusters
            silhouette_scores.append(silhouette_score(scaled_features, cluster_labels))
        else:
            silhouette_scores.append(-1) # Assign a low score if only one cluster is formed

    if silhouette_scores:
        optimal_clusters = K_range[silhouette_scores.index(max(silhouette_scores))]
        print(f"Optimal number of clusters (based on Silhouette Score): {optimal_clusters}")
    else:
        optimal_clusters = 2 # Default to 2 if no valid silhouette scores are found
        print("Could not determine optimal clusters using Silhouette Score, defaulting to 2.")

# Apply K-Means with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
customer_segmentation_df['Cluster'] = kmeans.fit_predict(scaled_features)

# 6. Display the head of customer_segmentation_df with the new 'Cluster' column
print("\nFirst 5 records of customer_segmentation_df with 'Cluster' column:")
display(customer_segmentation_df.head())

# Display the distribution of customers across the identified clusters
print("\nDistribution of customers across clusters:")
print(customer_segmentation_df['Cluster'].value_counts())

plt.figure(figsize=(8, 5))
sns.countplot(data=customer_segmentation_df, x='Cluster', hue='Cluster', palette='viridis', legend=False)
plt.title('Distribution of Customers Across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.show()
```

## Analyze Product Basket Analysis

### Subtask:
Apply association rule mining to identify frequently co-occurring items.


**Reasoning**:
To prepare the transaction data for market basket analysis, I need to import the `TransactionEncoder` and transform the list of items in the 'items' column of the `transactions` DataFrame into a one-hot encoded format. This is a prerequisite for applying association rule mining algorithms.



```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions['items']).transform(transactions['items'])
df_transactions = pd.DataFrame(te_ary, columns=te.columns_)

print("First 5 records of the one-hot encoded transactions DataFrame:")
display(df_transactions.head())
```

**Reasoning**:
With the one-hot encoded transaction data, I will now import the necessary functions from `mlxtend.frequent_patterns` to apply the Apriori algorithm to find frequent itemsets and then generate association rules. I will set the minimum support and lift thresholds as specified and display the top 10 rules.



```python
from mlxtend.frequent_patterns import apriori, association_rules

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_transactions, min_support=0.01, use_colnames=True)

print("First 5 frequent itemsets:")
display(frequent_itemsets.head())

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Sort rules by lift in descending order and display top 10
rules = rules.sort_values(by='lift', ascending=False)

print("\nTop 10 association rules by lift:")
display(rules.head(10))
```

**Reasoning**:
The previous attempt to generate association rules resulted in an empty DataFrame, likely due to a `min_support` threshold that was too high for the `apriori` algorithm. To find more frequent itemsets and subsequently more association rules, I will decrease the `min_support` to 0.005.



```python
from mlxtend.frequent_patterns import apriori, association_rules

# Apply Apriori algorithm to find frequent itemsets with a lower min_support
frequent_itemsets = apriori(df_transactions, min_support=0.005, use_colnames=True)

print("First 5 frequent itemsets (with min_support=0.005):")
display(frequent_itemsets.head())

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Sort rules by lift in descending order and display top 10
rules = rules.sort_values(by='lift', ascending=False)

print("\nTop 10 association rules by lift:")
display(rules.head(10))
```
